```python
con_data = pd.read_csv('control_group.csv',sep= ';')
test_data = pd.read_csv('test_group.csv', sep = ';')
```

```python
#lets change the weird names of our columns accross all data
```

```python
con_data.columns = ['Campaign Name', 'Date', 'Spend', 'Number of Impressions', 'Reach',
       'Website Clicks', 'Searches', 'View Content',
       'Add to Cart', 'Purchase']
```

```python
test_data.columns = ['Campaign Name', 'Date', 'Spend', 'Number of Impressions', 'Reach',
       'Website Clicks', 'Searches', 'View Content',
       'Add to Cart', 'Purchase']
```

```python
con_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
#let fill the missing values
```

```python
con_data['Number of Impressions'].fillna(value = con_data['Number of Impressions'].mean(),inplace = True)
con_data['Reach'].fillna(value= con_data['Reach'].mean(),inplace = True)
con_data['Website Clicks'].fillna(value = con_data['Website Clicks'].mean(),inplace = True)
con_data['Searches'].fillna(value = con_data['Searches'].mean(),inplace = True)
con_data['View Content'].fillna(value = con_data['View Content'].mean(),inplace = True)
con_data['Add to Cart'].fillna(value = con_data['Add to Cart'].mean(),inplace = True)
con_data['Purchase'].fillna(value = con_data['Purchase'].mean(),inplace = True)
```

```python
cons = sns.heatmap(data = con_data.isnull())
```

```python
con_data.head()
```

```python
con_data.info()
```

```python
test_data.info()
```

```python
# converting this set of float columns so i can merge them properly
con_data['Number of Impressions'] = con_data['Number of Impressions'].astype(int)
con_data['Reach'] = con_data['Reach'].astype(int)
con_data['Website Clicks'] = con_data['Website Clicks'].astype(int)
con_data['Searches'] = con_data['Searches'].astype(int)
con_data['View Content'] = con_data['View Content'].astype(int)
con_data['Add to Cart'] = con_data['Add to Cart'].astype(int)
con_data['Purchase'] = con_data['Purchase'].astype(int)
```

```python
con_data.info()
```

```python
AB_data = con_data.merge(test_data,how = 'outer').sort_values(['Date'])
AB_data = AB_data.reset_index(drop = True)
```
